```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data={'Name':['Ravi','Suresh','Ramesh','Sita','Gita','Hari','Shreya','anu','Rashmika','Pooja'],
      'Department':['IT','HR','IT','HR','IT','HR','IT','HR','IT','HR'],
      'Salary':[50000,60000,55000,65000,70000,75000,80000,85000,90000,None]}
df=pd.DataFrame(data)
avg=df['Salary'].mean()
df['Salary']=df['Salary'].fillna(avg)
print(df)

dept_salary=df.groupby('Department')['Salary'].mean().reset_index()
sns.barplot(x='Department',y='Salary',data=dept_salary)
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```
